```python
import pandas as pd
import numpy as np
import requests
import config
```

```python
# Read downloaded file
data = pd.read_csv('NetflixViewingHistory.csv')
data.head()
```

```python
data.info()
```

```python
# Change date column to date type 
data['Date'] = pd.to_datetime(data['Date'])
data
```

# Data Cleaning

```python
# Split column title
data[['Title','Season','Episode','Extra']] = data['Title'].str.split(':', 3,expand=True)
data
```

```python
data.drop(['Season','Episode','Extra'], axis=1, inplace = True)
```

```python
# Rename some titles
data['Title'] = data['Title'].replace({'Shameless (U.S.)':'Shameless',\
                                       'The Office (U.K.)':'The Office',\
                                       'The Office (U.S.)':'The Office',\
                                       'Kitchen Nightmares (U.S.)':'Kitchen Nightmares',\
                                       'Monarchy (U.K.)':'Monarchy',\
                                       'Mad Max 2':'The Road Warrior',\
                                       'Pocahontas II':'Pocahontas 2: Journey to a New World'                                    
                                     })
```

```python
# Retrieve unique title 
title = data['Title'].unique()
title
```

```python
# Make an api call to get language, country, rating info from omdi website
result = []
error = []

for t in range(0,len(title)):
    try:
        url = "http://www.omdbapi.com/?apikey="+config.api+"&t="+title[t]+"&plot=short&r=json"
        payload = {}
        headers= {}
        response = requests.request("GET", url, headers=headers, data = payload)
        print(response.text)
        result.append({'Title':response.json()['Title'],\
                   'Genre':response.json()['Genre'],\
                   'Language':response.json()['Language'],\
                   'Country':response.json()['Country'],\
                   'imdbRating':response.json()['imdbRating']})
    except:
        error.append(title[t])
```

```python
# Create a df
result_df = pd.DataFrame(result,columns=['Title','Genre','Language','Country','imdbRating'])
result_df
```

```python
# Merge data df and result df
merged_df = pd.merge(data, result_df, on='Title')
merged_df
```

```python
# Export csv file
merged_df.to_csv('Netflix Rating Output.csv')
```
